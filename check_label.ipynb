```python
import cv2
import matplotlib.pyplot as plt
import os
import numpy as np
import torch
import torchvision
import torch.nn as nn
```

```python
path = 'uploads/test/img/'
lbpath = 'uploads/test/label_cor/'
names = os.listdir(path)
np.random.shuffle(names)
for name in names:
    img = cv2.imread(path+name)[:,:,::-1]
    imcopy = img.copy()

    
    with open(lbpath+name[:-4] + '.txt', 'r') as f:
        label = f.readlines()
        label = np.array([[int(j) for j in i[:-1].split(' ')] for i in label])
    for lb in label:
        print(lb)
        cv2.circle(imcopy, lb, 7, (255,0,255), -1)
    plt.imshow(imcopy)

    break
```
